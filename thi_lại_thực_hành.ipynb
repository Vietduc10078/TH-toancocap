```python
from sympy import Integral, Symbol 
x = Symbol('x') 
Integral(2*x+1/x*(x+1), x).doit() 
```

```python
from sympy import Derivative,Symbol 
x=Symbol('x') 
f = (x**2+3*x-1)/(x+1) 
Derivative(f, x).doit() 
```

```python
from sympy import Derivative,Symbol 
x=Symbol('x') 
f = (x**2+3*x+3)/(x+1) 
Derivative(f, x).doit() 
```

```python
import sympy 
from sympy import *
x = Symbol('x') 
f = (sqrt(x**2-x+1)/(x+1))
c = Symbol ('c')
delta = Symbol('delta')
c = 2
delta = 1/4
sympy.plot (f,(x,c - delta,c + delta))
sympy.plot (f,abs(x),-abs(x),(x,c - delta,c + delta))

 
```
